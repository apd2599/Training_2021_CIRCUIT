Creating Outlier Filters using Record Blood Pressures

```python
lo_ap_hi_filter = (data["ap_hi"] < 0)
hi_ap_hi_filter = (data["ap_hi"] > 370)
lo_ap_lo_filter = (data["ap_lo"] < 0)
hi_ap_lo_filter = (data["ap_lo"] > 360)
```

```python
outlier_filter = (lo_ap_hi_filter | hi_ap_hi_filter | lo_ap_lo_filter | hi_ap_lo_filter)
outliers = data[~outlier_filter]
```

```python
data = data.reset_index().drop("index", axis = 1)
```

# Method 1: PCA/KMeans for Clustering 

PCA & KMeans

```python
model = KMeans(n_clusters=4)
model.fit(xs)
yhat = model.predict(xs)
clusters = unique(yhat)
```

```python
pca = PCA()
pca_data = pca.fit_transform(xs)
```

```python
pca_data0 = pca_data[yhat == 0]
pca_data1 = pca_data[yhat == 1]
pca_data2 = pca_data[yhat == 2]
pca_data3 = pca_data[yhat == 3]
```

```python
Clust0 = data_full_scaled[yhat == 0]
Clust1 = data_full_scaled[yhat == 1]
Clust2 = data_full_scaled[yhat == 2]
Clust3 = data_full_scaled[yhat == 3]
```

```python
cluster_labels = Clust0, Clust1, Clust2, Clust3
```

PCA Visualization

```python
plt.figure(figsize = [8,8])
plt.scatter(pca_data0[:, 0], pca_data0[:, 1], label = 0)
plt.scatter(pca_data1[:, 0], pca_data1[:, 1], label = 1)
plt.scatter(pca_data2[:, 0], pca_data2[:, 1], label = 2)
plt.scatter(pca_data3[:, 0], pca_data3[:, 1], label = 3)
plt.xlabel("PC 1 - 32 percent variance")
plt.ylabel("PC 2 - 21 percent variance")
plt.legend()
```

There are 4 Clusters (Green, Purple, Blue, and Yellow) as seen by the different colors. Along the x-axis, the clustering seems to be binary as there are clusters found near (-0.6, 0) and (0.4, 1). This suggests that one of the more predominant separating factors along the x-axis is binary. Along the y-axis, there are also two distinct groups (green/purple and yellow/blue). However, here there is more separation as there 4 lines at different y-values within each cluster. This suggests that there are both a binary and continuous variable that affects the clustering along the y-axis. Overall, the clustering seen above does a really good job of sorting the data into the 4 distinct groups, there doesn't seem to be any apparent inaccuracies based on the colors of the clusters in the visual.

```python
pca.explained_variance_ratio_[1]
```

```python
#pca.components_[:,0]
pd.Series(pca.components_[:,0],index = xs.columns).abs().sort_values(ascending = False)
```

The Series above shows the list of the features in descending order of variance/breadth. The glucose level shows to account for the majority of the variance, suggesting that the patients are falling along varying levels of glucose. The glucose feature has three potential outcomes (1,2,3) in order of worsening glucose levels beginning with normal levels at 1. This suggests that glucose would be the main separating factor along the x-axis, potentially into normal glucose levels as one side and the other could be above/well above normal levels. Additionally, age accounts for a small amount of the variance along the x-axis, which could account for the spread of the two different clustering halves, due to the continuity of the age factor.

```python
#pca.components_[:,1]
pd.Series(pca.components_[:,1],index = xs.columns).abs().sort_values(ascending = False)
```

The Series above shows the list of the features in descending order of variance/breadth. The gender feature shows to account for nearly all of the variance along the y-axis, suggesting that the gender feature seems to be important to separating patients into clusters. The gender feature has binary outcomes, with 1 for female and 2 for male. Additionally, systolic blood pressure accounts for approximately 3% of the variance along the y-axis, which could account for the difference in the spread among the clusters, due to the continuity of the systolic blood pressure.
